```python
# Import package
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Specify url: url
res = []
url = 'https://www.the-numbers.com/movie/budgets/all'
urllist = [url]
# urllist = []
numofiter = 59
```

```python
def urlAppend(urllist,url):
    return urllist.append(url) 
  
def genUrl(url,b):
    return url+"/"+str(b) 
```

```python
j=1
for i in range(1,numofiter):
    j+=100
    s=genUrl(url,j)
    urlAppend(urllist,s)
    print(s) 
    
```

```python
    
df1 = pd.DataFrame(urllist)    
print(df1.head(200))
```

```python
# Package the request, send the request and catch the response
from random import randint
from time import sleep

docs = []
for url in urllist:
    print(url)
    r = requests.get(url)
    sleep(randint(1,10))
# Extract the response as html: html_doc
    html_doc = r.text
    urlAppend(docs,html_doc)
# print(docs)
```

```python
res1 = []
for doc in docs:    
    #Create a BeautifulSoup object from the HTML: soup
    soup = BeautifulSoup(doc)
# Get the title of the webpage
    title = soup.title
# Print the title of Guido's webpage to the shell
#     print(title)

    for tr in soup.find_all('tr')[1:]:
        tds = tr.find_all('td')
#     row = ((tds[0].text, tds[1].text,tds[2].text, tds[3].text,tds[4].text,tds[5].text))
        row = [trs.text.strip() for trs in tds if trs.text.strip()]
        if row:
            title_href=tds[2].find('a')
            if title_href: row.append(title_href.get('href'))      
            else: row.append('')
            row.append(doc)
            res1.append(row)
                      
df = pd.DataFrame(res1, columns=['No.','ReleaseDate','MovieName','ProductionBudget','DomesticGross','WorldwideGross', 'href','doc'])
print(df.head)




                       
```
