```python

def connect(db, user, password, host="localhost", port=5432):
    """
    Connects to the database and ensures there are tables.
    """
    conn = psycopg2.connect(**{
        "database": db,
        "user": user,
        "password": password,
        "host": host,
        "port": port
    })
    return conn
                
```
