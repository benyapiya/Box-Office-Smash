```python
moviepage = requests.get("https://www.the-numbers.com/movie/Avatar#tab=summary")
moviesoup = BeautifulSoup(moviepage.content,'html.parser')
```

```python
print(moviesoup.prettify())
```

```python
kwsoup = moviesoup.find_all('a', href=re.compile('.keywords.'))
kwlist = [tag.contents[0] for tag in kwsoup]
print(kwlist)
```
